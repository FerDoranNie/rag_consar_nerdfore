```python
from google_auth_oauthlib.flow import InstalledAppFlow
import os
import logging
```

```python
## funciones locales

logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

SCOPES = ['https://www.googleapis.com/auth/drive']
SCOPES = ['https://www.googleapis.com/auth/drive.file']

CREDENTIALS_FILE = '../account_keys_secrets/credentials_oauth.json'
TOKEN_FILE = 'token.json'
FIXED_PORT = 8080
TIMEOUT_SECONDS = 60

try:
    if not os.path.exists(CREDENTIALS_FILE):
        raise FileNotFoundError(f"El archivo de credenciales no se encontró en: {CREDENTIALS_FILE}")

    flow = InstalledAppFlow.from_client_secrets_file(CREDENTIALS_FILE, SCOPES)
    
    logging.info(f"Iniciando servidor local en el puerto {FIXED_PORT}. Por favor, autoriza la aplicación en tu navegador.")
    logging.info(f"El proceso se cancelará automáticamente en {TIMEOUT_SECONDS} segundos si no se completa la autenticación.")

    creds = flow.run_local_server(port=FIXED_PORT, timeout_seconds=TIMEOUT_SECONDS)

    # Guardar token para futuras ejecuciones
    with open(TOKEN_FILE, 'w') as token:
        token.write(creds.to_json())

    logging.info(f"Token generado y guardado exitosamente en {TOKEN_FILE}")

except FileNotFoundError as e:
    logging.error(e)
except (OSError, TimeoutError) as e:
    logging.error(f"Error: Se agotó el tiempo de espera de {TIMEOUT_SECONDS} segundos. No se pudo completar la autenticación.")
    logging.error(f"Detalle del error: {e}")
except Exception as e:
    logging.error(f"Ocurrió un error inesperado durante el flujo de autenticación: {e}")

```
